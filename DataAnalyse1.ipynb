```python
def numberPoint(x , y):
    score = 0

  for x ,y in foot_data
    if (x > y):
        score +=3
    elif (x < y):
        score -=3
    else :
        score +=1
    return score
```

```python
foot_data['PointGagne'] = foot_data.apply((numberPoint(foot_data['home_score'],foot_data['away_score'])))
```

```python
foot_data['date']= foot_data['date'].map(pd.to_datetime)
```

```python
def get_year(dt):
    return dt.year
foot_data['year']= foot_data['date'].map(get_year)

def get_month(dt):
    return dt.month
foot_data['month']= foot_data['date'].map(get_month)

def get_day(dt):
    return dt.day
foot_data['day']= foot_data['date'].map(get_day)

def get_weekday(dt):
    return dt.day
foot_data['weekday']= foot_data['date'].map(get_weekday)

foot_data.head()
```

```python
hist(foot_data.day, bins=50, rwidth=.5, range=(0.5, 30.5))
xlabel('date of the month')
ylabel('frequency')
title('Frequency by Day')
```

```python
hist(foot_data.weekday, bins =7, range=(-.5, 6.5), rwidth=.8, color='red' , alpha=.4)
xticks(range(7), 'Mon Tue Wed Thu Fri Sat Sun'.split());
```

```python
hist(foot_data.month, bins=80, rwidth=.8 , range=(0.5, 12.5))
xlabel('month of the periode')
ylabel('frequency')
title('Frequency by Month')
```
